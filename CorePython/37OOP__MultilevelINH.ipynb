```python
class Device:
  def power_on(self):
    print("Device powered ON")
  def power_off(self):
    print("Device powered OFF")
```

```python
class SmartDevice(Device):
  def connect_wifi(self):
    print("connected to wifi..SohamFiber")
  def install_app(self):
    print("New app installed...")
```

```python
class SmartWatch(SmartDevice):
  def track_steps(self):
    print("counting steps..you keep walking")
  def heart_rate(self):
    print("heart rate measurement is on")
```

```python
obj=SmartWatch()
obj.power_on()
obj.connect_wifi()
obj.install_app()
obj.track_steps()
obj.heart_rate()
obj.power_off()
```
