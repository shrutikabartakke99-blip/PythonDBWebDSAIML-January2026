```python
class Calculator:
  def calcsquare(self,n):
    sq=n**2
    print(f"square of {n} is {sq}")
  def multiply(self,x,y):
    res=x*y
    print(f"multiplication of {x} and {y} is {res}")
  def add(self,x,y):
    res=x+y
    print(f"addition of {x} and {y} is {res}")
```

```python
obj=Calculator()
```

```python
obj.multiply(12,23)
```

```python
obj.add(12,23)
```

```python
obj.calcsquare(12)
```
