```python
s="madam"
s1=s[::-1]

if s==s1:
  print("palindrome")
else:
  print("not a palindrome")
```

```python
s="technology is power"
vowels="aeiouAEIOU"
count=0

for ch in s:
  if ch in vowels:
    count+=1
print(f"number of vowels in the string are {count}")
```

```python
s1="listen"
s2="silent"

print(sorted(s1))

if sorted(s1)==sorted(s2):
  print("anagram")
else:
  print("not anagram")
```
