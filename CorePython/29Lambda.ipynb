Lambda function is a small, single line, anonymous function.

```python
def calcsquare(x):
  sq=x**2
  return sq
```

```python
print("square is ",calcsquare(15))
```

lambda parameters:return value

```python
sq=lambda x:x**2
```

```python
print('square is ',sq(5))
```

```python
print(sq(-9))
```

```python
add=lambda a,b:a+b
```

```python
print(add(4,8))
```

```python
d=lambda amount:amount*10/100
```

```python
print(d(3400))
```
