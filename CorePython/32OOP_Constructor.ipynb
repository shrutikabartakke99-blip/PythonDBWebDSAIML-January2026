Constructor is a special function in the class that runs automatically when the object is created to perform some initialization activities

```python
class Customer:
  def __init__(self):
    print("Welcome to ApnaBazar shopping mall")

  def calcdiscount(self,amount):
    if amount>20000:
      print('13% discount applied')
      discount=amount*13/100
    else:
      print('9% discount applied')
      discount=amount*9/100
    print(f'discount on {amount} is {discount}')
```

```python
# when object is created constructor runs automatically
obj=Customer()
```

```python
amt=float(input('Enter purchase amount : '))
obj.calcdiscount(amt)
```
