```python
class User:
  def login(self):
    print("login successful...")
```

```python
class Customer(User):
  def order_food(self):
    print("Food ordered & will be delivered in 30 minutes")
```

```python
class DeliveryBoy(User):
  def deliver(self):
    print("Order delivered in time")
```

```python
c=Customer()
c.login()   # inherited method
c.order_food()  # own function
```

```python
d=DeliveryBoy()
d.login()
d.deliver()
```
