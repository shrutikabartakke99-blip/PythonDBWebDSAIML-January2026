find out if the number is odd or even

```python
n=int(input('Enter a number : '))
if n%2==0:
  print('number is even')
else:
  print('number is odd')
```

```python
n=int(input('Enter a number : '))
if n==0:
  print(f'{n} is a neutral number')
else:
  if n>0:
    print('number is positive')
  else:
    print('number is negative')
```

```python
custnm=input('Enter customer name : ')
amount=float(input('Enter purchase amount : '))
if amount>20000:
  print('13% discount applied')
  discount=amount*13/100
else:
  print('9% discount applied')
  discount=amount*9/100

netbill=amount-discount
print(f'Customer Name : {custnm}')
print(f'Amount : {amount}')
print(f'Discount : {discount}')
print(f'Net Bill : {netbill}')
```
